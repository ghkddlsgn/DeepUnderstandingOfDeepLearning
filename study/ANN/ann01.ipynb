```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from IPython import display
```

```python
N = 30
torch.manual_seed(42)
x = torch.randn(N, 1)
y = x + torch.randn(N,1)

plt.plot(x,y,'s')
plt.show()
```

```python
ANNreg = nn.Sequential(
    nn.Linear(1,1), #1 input, 1 output for input layer
    nn.ReLU(),
    nn.Linear(1,1) 
)

ANNreg
```

```python
learningRate = .05
lossfun = nn.MSELoss()
optimizer = torch.optim.SGD(ANNreg.parameters(), lr=learningRate) #using gradient descient
```

```python
numepochs = 500
losses = torch.zeros(numepochs)

for epochi in range(numepochs):
    
    yHat = ANNreg(x)

    loss = lossfun(yHat, y)
    losses[epochi] = loss

    optimizer.zero_grad()
    loss.backward()
    optimizer.step
```
